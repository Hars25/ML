```python
#Merging match_df with delivery df
delivery_df=match_df.merge(deliveries,on="match_id")
```

```python
delivery_df.head()
```

```python
delivery_df=delivery_df[delivery_df['inning']==2]
```

```python
delivery_df.head()
```

```python
# Current score of particular match
delivery_df['current_score']=delivery_df.groupby("match_id").cumsum()['total_runs_y']
```

```python
delivery_df.head()
```

```python
delivery_df['runs_left']=delivery_df['total_runs_x']-delivery_df['current_score']
```

```python
delivery_df.head()
```

```python
delivery_df['balls_left']=126-(delivery_df['over']*6+delivery_df['ball'])
```

```python
delivery_df.head()
```

```python
delivery_df['player_dismissed'].unique()
```

```python
#here player_dismissed means 0 else 1
delivery_df['player_dismissed']=delivery_df['player_dismissed'].fillna(0)
```

```python
delivery_df['player_dismissed']=delivery_df['player_dismissed'].apply(lambda x:x if x==0 else 1)
```

```python
delivery_df['player_dismissed'].unique()
```

```python
#wickets_fallen
wickets=delivery_df.groupby("match_id").cumsum()['player_dismissed'].values
wickets
```

```python
delivery_df['wickets']=10-wickets
```

```python
delivery_df.head()
```

```python
#Current Run rate
delivery_df['cur_run_rate']=(delivery_df['current_score']*6)/(120-delivery_df['balls_left'])
```

```python
delivery_df.head()
```

```python
#Required Run rate
delivery_df['req_run_rate']=(delivery_df['runs_left']*6)/(delivery_df['balls_left'])
```

```python
delivery_df.head()
```

```python
def resultfun(row):
    return 1 if row['batting_team']==row['winner'] else 0
```

```python
delivery_df['result']=delivery_df.apply(resultfun,axis=1)
```

```python
delivery_df.head()
```

```python
final_df=delivery_df[['batting_team','bowling_team','city','total_runs_x','runs_left','balls_left','wickets','cur_run_rate','req_run_rate','result']]
```

```python
final_df.head()
```

```python
final_df=final_df.dropna()
```

```python
final_df=final_df[final_df['balls_left']!=0]
```

```python
final_df.head()
```

```python
X=final_df.drop(['result'],axis=1)
y=final_df['result']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=123)
```

```python
X_train.shape,y_train.shape
```

```python
X_test.shape,y_test.shape
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn import metrics
```

```python
#cat_col
cat_col=final_df.select_dtypes(include=['O']).columns
cat_col
```

```python
cf=ColumnTransformer([('trf',OneHotEncoder(sparse=False,drop="first"),['batting_team','bowling_team','city'])],remainder="passthrough")
```

```python
pipe=Pipeline(steps=[('step1',cf),('step2',LogisticRegression(solver="liblinear"))])
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred=pipe.predict(X_test)
```

```python
pipe.predict_proba(X_test)[10]
```

```python
pipe1.fit(X_train,y_train)
```
